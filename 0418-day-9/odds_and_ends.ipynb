# 🐍 **PYTHON "ODDS AND ENDS"**
This notebook covers Python's expressive and idiomatic features—those little tools that make Python powerful and elegant. See definitions and examples below!

### **LIST COMPREHENSION**
List comprehension is a concise way to generate a new list by applying an expression to each item in an iterable. It can optionally include an `if` clause to filter items.

### Syntax
```python
[expression for item in iterable if condition]
```

### Example
```python
squares = [x*x for x in range(10)]
evens = [x for x in range(20) if x % 2 == 0]
print(squares)
print(evens)
```

### **TERNARY OPERATOR**
A ternary operator is a shorthand for an if-else statement that returns a value based on a condition.

### Syntax
```python
a if condition else b
```

### Example
```python
score = 75
status = "pass" if score >= 60 else "fail"
print(status)
```

### **MULTIPLE ASSIGNMENT AND UNPACKING**
Python allows you to assign multiple variables at once and unpack values from lists/tuples.

### Examples
```python
a, b, c = [1, 2, 3]     # multiple assignment
a, b = b, a             # swap values
head, *body, tail = [1, 2, 3, 4, 5]  # unpack with *
print(head, body, tail)
```

### **ENUMERATE**
`enumerate()` adds a counter to an iterable and returns it as an enumerate object, commonly used in loops.

### Example
```python
for i, val in enumerate(["a", "b", "c"]):
    print(i, val)
```

### **ZIP**
`zip()` combines multiple iterables element-wise into tuples. Useful for parallel iteration.

### Example
```python
names = ["Alice", "Bob"]
ages = [25, 30]
for name, age in zip(names, ages):
    print(f"{name} is {age}")
```

### **F-STRINGS**
F-strings (formatted string literals) provide a clean syntax for embedding expressions inside string literals.

### Example
```python
name = "Mike"
print(f"Hello, {name.upper()}!")
```

### **ANY/ALL**
- `any(iterable)` returns True if any element is truthy.
- `all(iterable)` returns True only if all elements are truthy.

### Example
```python
nums = [1, 2, 3, 0]
print(any(nums))  # True, because some are non-zero
print(all(nums))  # False, because of the zero
```

### **LAMBDA**
`lambda` creates small anonymous functions inline, often used in `map`, `filter`, etc.

### Example
```python
double = lambda x: x * 2
print(double(4))  # 8
```

### **MAP/FILTER**
- `map(func, iterable)` applies a function to every item.
- `filter(func, iterable)` filters items based on a function that returns True/False.

### Example
```python
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
evens = list(filter(lambda x: x % 2 == 0, nums))
print(squared)
print(evens)
```

### **SET/DICT COMPREHENSIONS**
Just like list comprehensions, but creates sets or dictionaries.

### Example
```python
unique_squares = {x**2 for x in range(5)}
squares_dict = {x: x**2 for x in range(5)}
print(unique_squares)
print(squares_dict)
```

### **UNDERSCORE/DUMMY VARIABLES**
The `_` is often used as a throwaway variable in loops or pattern unpacking.

### Example
```python
for _ in range(3): print("Hi")
```

### **WALRUS OPERATOR**
Introduced in Python 3.8, the walrus operator `:=` allows assignment inside expressions.

### Example
```python
data = [1, 2, 3]
if (n := len(data)) > 2:
    print(f"Length is {n}")
```

### **SLICING**
Extract a portion of a list using `[start:stop:step]`. Supports negative indices and steps.

### Example
```python
lst = [0, 1, 2, 3, 4, 5]
print(lst[::2])   # [0, 2, 4]
print(lst[::-1])  # reversed list
```

### **CHAINED COMPARISONS**
Python allows chained comparisons, like in math, for readability.

### Example
```python
x = 5
if 0 < x < 10:
    print("x is in range")
```

### **DEFAULTDICT/COUNTER**
From `collections` module:
- `defaultdict` automatically initializes missing keys.
- `Counter` counts occurrences.

### Example
```python
from collections import defaultdict, Counter
dd = defaultdict(int)
dd['a'] += 1
print(dd)
freq = Counter("banana")
print(freq)
```

### **TRY/EXCEPT/ELSE/FINALLY**
Error handling in Python:
- `try` runs code
- `except` handles exceptions
- `else` runs if no exceptions
- `finally` always runs

### Example
```python
x = 1
try:
    result = 1 / x
except ZeroDivisionError:
    result = 0
else:
    print("No error")
finally:
    print("Done")
```

### **\*ARGS AND **KWARGS**
- `*args` lets you pass variable positional arguments.
- `**kwargs` lets you pass variable keyword arguments.

### Example
```python
def f(*args, **kwargs):
    print(args, kwargs)
f(1, 2, a=3, b=4)
```

### **TYPE HINTS**
Optional annotations to indicate expected types of variables, arguments, or return values.

### Example
```python
def greet(name: str) -> str:
    return f"Hello, {name}"
print(greet("Alice"))
```

### **LIST FLATTENING**
You can flatten nested lists using a nested list comprehension.

### Example
```python
matrix = [[1, 2], [3, 4]]
flat = [item for row in matrix for item in row]
print(flat)
```
